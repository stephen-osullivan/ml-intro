### This notebook walks through fitting models to the IRIS dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)
print(iris.keys())
data = iris['data']
data['label'] = pd.Categorical(iris['target'])
data
```

```python
data['is_train'] = np.random.uniform(size=len(data)) < 0.66
data['is_train'].value_counts()
```

```python
g = sns.pairplot(data, hue='label', height=2, plot_kws={"alpha":0.5})
plt.show()
```
